```
import warnings
warnings.filterwarnings("ignore")

# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from newspaper import Article
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```
#Scarping data dari wikipedi dengan menggunakan artiket berjudul "Cokelat"
from newspaper import Article
article = Article("https://en.wikipedia.org/wiki/Cokelat")
article.download()
article.parse()
```

```
wc = WordCloud(background_color="white", max_words=2000,
               stopwords=STOPWORDS, max_font_size=256,
               random_state=42, width=500, height=500)
wc.generate(article.text)
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```
# Bentuk bunga
mask = np.array(Image.open('bung.jpg'))
wc = WordCloud(stopwords=STOPWORDS, mask=mask, background_color="white",
               max_words=2000, max_font_size=256,
               random_state=42, width=mask.shape[1],
               height=mask.shape[0])
wc.generate(article.text)
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```
# Gambar Object yaitu Cartoon Patrick
mask1 = np.array(Image.open('PatrickStar.jpg'))
mask_colors = ImageColorGenerator(mask1)
wc = WordCloud(stopwords=STOPWORDS, mask=mask1, background_color="white",
               max_words=2000, max_font_size=256,
               random_state=42, width=mask1.shape[1],
               height=mask1.shape[0], color_func=mask_colors)
wc.generate(article.text)
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```
