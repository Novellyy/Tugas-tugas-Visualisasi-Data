#### Code

```python
#Hang_Seng
Hang_df = pd.read_csv("Hang_Seng.csv") 
Hang_df.head()
```

```python
years19 = (Hang_df['Date'] > '2018-12-31') & (Hang_df['Date'] < '2020-01-01')
```

```python
Hang_df = Hang_df.loc[years19]
Hang_df.head()
```

```python
Hang_df["Day_Perc_Change"] = Hang_df["Adj Close"].pct_change()*100
Hang_df.head()
```

```python
Hang_df.dropna(axis = 0, inplace = True)
```

```python
def trend(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"
    
# HDFC_df["Trend"]= np.zeros(HDFC_df["Day_Perc_Change"].count())
Hang_df["Trend"]= Hang_df["Day_Perc_Change"].apply(lambda x:trend(x))
Hang_df.head()
```

```python
Hang_pie_data = Hang_df.groupby('Trend')
pie_label = sorted([i for i in Hang_df.loc[:, 'Trend'].unique()])
plt.pie(Hang_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)

plt.show()
```

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
# load stock market data

Hang_df = pd.read_csv("Hang_Seng.csv") 
Nas_df = pd.read_csv("Nasdaq.csv") 
Nik_df = pd.read_csv("Nikkei.csv") 

# merge Adj. close values
combined_df = pd.concat([Hang_df["Adj Close"], Nas_df["Adj Close"], Nik_df["Adj Close"]], axis=1)
combined_df.columns = ["Hang_Seng", "Nasdaq", "Nikkei"]
```

```python
# store daily returns of all above stocks in a new dataframe 
pct_chg_df = combined_df.pct_change()*100
pct_chg_df.dropna(inplace = True, how = "any", axis = 0)
```

```python
# type your code here
Hang_vol = pct_chg_df["Hang_Seng"].rolling(7).std()*np.sqrt(7)
Hang_vol.plot(figsize = (15, 7))
```

```python
# type your code here
Nas_vol = pct_chg_df["Nasdaq"].rolling(7).std()*np.sqrt(7)
Nas_vol.plot(figsize = (15, 7))
```

```python
# type your code here
Nik_vol = pct_chg_df["Nikkei"].rolling(7).std()*np.sqrt(7)
Nik_vol.plot(figsize = (15, 7))
```

```python
volatility = pct_chg_df[['Hang_Seng', 'Nasdaq', 'Nikkei']].rolling(7).std()*np.sqrt(7)
volatility.plot(figsize = (15, 7))
```

#### Penjelasan 
Pada ketiga saham ini, sama-sama memiliki pergerakan yang dapat dikatakan tinggi/Volatilitas yang tinggi dikarenakan pada grafik terdapat pergerakan yang fluktuatif. Seperti pada saham Nikkei yang mengalami fluktuatif yang tinggi pada bulan 3 sampai bulan ke 5, saham Nasdaq yang mengalami fluktuatif dari bulan 8 sampai bulan 9, dan saham Hang_Seng yang dikategorikan unggul jika dibandingkan dengan saham Nikkei dan saham Nasdaq.
